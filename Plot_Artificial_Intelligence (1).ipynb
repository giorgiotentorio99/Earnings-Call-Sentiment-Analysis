```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline

model_name = "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

def split_text(text, max_len=512):
    sentences = text.split('.')
    chunks, chunk = [], ""
    for sentence in sentences:
        if len(chunk) + len(sentence) < max_len:
            chunk += sentence + "."
        else:
            chunks.append(chunk)
            chunk = sentence + "."
    if chunk:
        chunks.append(chunk)
    return chunks
def analyze_sentiment(text):
    chunks = split_text(text)
    results = sentiment_pipeline(chunks)
    sentiment_scores = {"LABEL_0": 0, "LABEL_1": 0, "LABEL_2": 0}
    for res in results:
        sentiment_scores[res['label']] += res['score']

    total_score = sum(sentiment_scores.values())

    positive_percentage = (sentiment_scores["LABEL_2"] / total_score) * 100

    return positive_percentage

microsoft_sentiment120 = analyze_sentiment(microsoft_text1_20)
microsoft_sentiment220 = analyze_sentiment(microsoft_text2_20)
microsoft_sentiment320 = analyze_sentiment(microsoft_text3_20)
microsoft_sentiment420 = analyze_sentiment(microsoft_text4_20)
microsoft_sentiment121 = analyze_sentiment(microsoft_text1_21)
microsoft_sentiment221 = analyze_sentiment(microsoft_text2_21)
microsoft_sentiment321 = analyze_sentiment(microsoft_text3_21)
microsoft_sentiment421 = analyze_sentiment(microsoft_text4_21)
microsoft_sentiment122 = analyze_sentiment(microsoft_text1_22)
microsoft_sentiment222 = analyze_sentiment(microsoft_text2_22)
microsoft_sentiment322 = analyze_sentiment(microsoft_text3_22)
microsoft_sentiment422 = analyze_sentiment(microsoft_text4_22)
microsoft_sentiment123 = analyze_sentiment(microsoft_text1_23)
microsoft_sentiment223 = analyze_sentiment(microsoft_text2_23)
microsoft_sentiment323 = analyze_sentiment(microsoft_text3_23)
microsoft_sentiment423 = analyze_sentiment(microsoft_text4_23)
microsoft_sentiment124 = analyze_sentiment(microsoft_text1_24)
microsoft_sentiment224 = analyze_sentiment(microsoft_text2_24)
microsoft_sentiment324 = analyze_sentiment(microsoft_text3_24)
microsoft_sentiment424 = analyze_sentiment(microsoft_text4_24)
tesla_sentiment120 = analyze_sentiment(tesla_text1_20)
tesla_sentiment220 = analyze_sentiment(tesla_text2_20)
tesla_sentiment320 = analyze_sentiment(tesla_text3_20)
tesla_sentiment420 = analyze_sentiment(tesla_text4_20)
tesla_sentiment121 = analyze_sentiment(tesla_text1_21)
tesla_sentiment221 = analyze_sentiment(tesla_text2_21)
tesla_sentiment321 = analyze_sentiment(tesla_text3_21)
tesla_sentiment421 = analyze_sentiment(tesla_text4_21)
tesla_sentiment122 = analyze_sentiment(tesla_text1_22)
tesla_sentiment222 = analyze_sentiment(tesla_text2_22)
tesla_sentiment322 = analyze_sentiment(tesla_text3_22)
tesla_sentiment422 = analyze_sentiment(tesla_text4_22)
tesla_sentiment123 = analyze_sentiment(tesla_text1_23)
tesla_sentiment223 = analyze_sentiment(tesla_text2_23)
tesla_sentiment323 = analyze_sentiment(tesla_text3_23)
tesla_sentiment423 = analyze_sentiment(tesla_text4_23)
tesla_sentiment124 = analyze_sentiment(tesla_text1_24)
tesla_sentiment224 = analyze_sentiment(tesla_text2_24)
tesla_sentiment324 = analyze_sentiment(tesla_text3_24)
tesla_sentiment424 = analyze_sentiment(tesla_text4_24)
apple_sentiment120 = analyze_sentiment(apple_text1_20)
apple_sentiment220 = analyze_sentiment(apple_text2_20)
apple_sentiment320 = analyze_sentiment(apple_text3_20)
apple_sentiment420 = analyze_sentiment(apple_text4_20)
apple_sentiment121 = analyze_sentiment(apple_text1_21)
apple_sentiment221 = analyze_sentiment(apple_text2_21)
apple_sentiment321 = analyze_sentiment(apple_text3_21)
apple_sentiment421 = analyze_sentiment(apple_text4_21)
apple_sentiment122 = analyze_sentiment(apple_text1_22)
apple_sentiment222 = analyze_sentiment(apple_text2_22)
apple_sentiment322 = analyze_sentiment(apple_text3_22)
apple_sentiment422 = analyze_sentiment(apple_text4_22)
apple_sentiment123 = analyze_sentiment(apple_text1_23)
apple_sentiment223 = analyze_sentiment(apple_text2_23)
apple_sentiment323 = analyze_sentiment(apple_text3_23)
apple_sentiment423 = analyze_sentiment(apple_text4_23)
apple_sentiment124 = analyze_sentiment(apple_text1_24)
apple_sentiment224 = analyze_sentiment(apple_text2_24)
apple_sentiment324 = analyze_sentiment(apple_text3_24)
apple_sentiment424 = analyze_sentiment(apple_text4_24)

print("Microsoft Sentiment Score Q1 2020:", microsoft_sentiment120)
print("Microsoft Sentiment Score Q2 2020:", microsoft_sentiment220)
print("Microsoft Sentiment Score Q3 2020:", microsoft_sentiment320)
print("Microsoft Sentiment Score Q4 2020:", microsoft_sentiment420)

print("Microsoft Sentiment Score Q1 2021:", microsoft_sentiment121)
print("Microsoft Sentiment Score Q2 2021:", microsoft_sentiment221)
print("Microsoft Sentiment Score Q3 2021:", microsoft_sentiment321)
print("Microsoft Sentiment Score Q4 2021:", microsoft_sentiment421)

print("Microsoft Sentiment Score Q1 2022:", microsoft_sentiment122)
print("Microsoft Sentiment Score Q2 2022:", microsoft_sentiment222)
print("Microsoft Sentiment Score Q3 2022:", microsoft_sentiment322)
print("Microsoft Sentiment Score Q4 2022:", microsoft_sentiment422)

print("Microsoft Sentiment Score Q1 2023:", microsoft_sentiment123)
print("Microsoft Sentiment Score Q2 2023:", microsoft_sentiment223)
print("Microsoft Sentiment Score Q3 2023:", microsoft_sentiment323)
print("Microsoft Sentiment Score Q4 2023:", microsoft_sentiment423)

print("Microsoft Sentiment Score Q1 2024:", microsoft_sentiment124)
print("Microsoft Sentiment Score Q2 2024:", microsoft_sentiment224)
print("Microsoft Sentiment Score Q3 2024:", microsoft_sentiment324)
print("Microsoft Sentiment Score Q4 2024:", microsoft_sentiment424)

print("Tesla Sentiment Score Q1 2020:", tesla_sentiment120)
print("Tesla Sentiment Score Q2 2020:", tesla_sentiment220)
print("Tesla Sentiment Score Q3 2020:", tesla_sentiment320)
print("Tesla Sentiment Score Q4 2020:", tesla_sentiment420)

print("Tesla Sentiment Score Q1 2021:", tesla_sentiment121)
print("Tesla Sentiment Score Q2 2021:", tesla_sentiment221)
print("Tesla Sentiment Score Q3 2021:", tesla_sentiment321)
print("Tesla Sentiment Score Q4 2021:", tesla_sentiment421)

print("Tesla Sentiment Score Q1 2022:", tesla_sentiment122)
print("Tesla Sentiment Score Q2 2022:", tesla_sentiment222)
print("Tesla Sentiment Score Q3 2022:", tesla_sentiment322)
print("Tesla Sentiment Score Q4 2022:", tesla_sentiment422)

print("Tesla Sentiment Score Q1 2023:", tesla_sentiment123)
print("Tesla Sentiment Score Q2 2023:", tesla_sentiment223)
print("Tesla Sentiment Score Q3 2023:", tesla_sentiment323)
print("Tesla Sentiment Score Q4 2023:", tesla_sentiment423)

print("Tesla Sentiment Score Q1 2024:", tesla_sentiment124)
print("Tesla Sentiment Score Q2 2024:", tesla_sentiment224)
print("Tesla Sentiment Score Q3 2024:", tesla_sentiment324)
print("Tesla Sentiment Score Q4 2024:", tesla_sentiment424)

print("Apple Sentiment Score Q1 2020:", apple_sentiment120)
print("Apple Sentiment Score Q2 2020:", apple_sentiment220)
print("Apple Sentiment Score Q3 2020:", apple_sentiment320)
print("Apple Sentiment Score Q4 2020:", apple_sentiment420)

print("Apple Sentiment Score Q1 2021:", apple_sentiment121)
print("Apple Sentiment Score Q2 2021:", apple_sentiment221)
print("Apple Sentiment Score Q3 2021:", apple_sentiment321)
print("Apple Sentiment Score Q4 2021:", apple_sentiment421)

print("Apple Sentiment Score Q1 2022:", apple_sentiment122)
print("Apple Sentiment Score Q2 2022:", apple_sentiment222)
print("Apple Sentiment Score Q3 2022:", apple_sentiment322)
print("Apple Sentiment Score Q4 2022:", apple_sentiment422)

print("Apple Sentiment Score Q1 2023:", apple_sentiment123)
print("Apple Sentiment Score Q2 2023:", apple_sentiment223)
print("Apple Sentiment Score Q3 2023:", apple_sentiment323)
print("Apple Sentiment Score Q4 2023:", apple_sentiment423)

print("Apple Sentiment Score Q1 2024:", apple_sentiment124)
print("Apple Sentiment Score Q2 2024:", apple_sentiment224)
print("Apple Sentiment Score Q3 2024:", apple_sentiment324)
print("Apple Sentiment Score Q4 2024:", apple_sentiment424)

```

```python
!pip install numpy
!pip install pandas
!pip install scipy
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats

microsoft_prices = pd.read_csv('Microsoft Stock Price History (2).csv', index_col='Date', parse_dates=True)
tesla_prices = pd.read_csv('Tesla Stock Price History (2).csv', index_col='Date', parse_dates=True)
apple_prices = pd.read_csv('Apple Stock Price History (2).csv', index_col='Date', parse_dates=True)
microsoft_sentiment = [
    90.22, 89.95, 81.36, 90.01,
    87.83, 90.30, 93.37, 89.26,
    88.24, 86.80, 74.74, 66.33,
    81.41, 64.46, 86.66, 79.09,
    74.53, 74.53, 79.98, 75.85
]

tesla_sentiment= [
    66.51, 70.76, 70.78, 60.77,
    51.20, 48.07, 61.10, 49.31,
    59.40, 65.92, 70.00, 66.27,
    58.39, 62.09, 43.93, 66.99,
    56.94, 57.36, 68.14, 49.95
]

apple_sentiment = [
    85.16, 85.16, 88.75, 84.69,
    84.23, 91.44, 92.34, 80.76,
    89.26, 78.11, 77.71, 85.31,
    84.52, 92.44, 92.17, 84.41,
    93.76, 91.40, 87.24, 92.24
]

microsoft_speech_dates = ['2020-01-29', '2020-04-29', '2020-07-22','2020-10-27', '2021-01-26', '2021-04-27', '2021-07-27','2021-10-26', '2022-01-25', '2022-04-26', '2022-07-27','2022-10-25', '2023-01-24', '2023-04-25', '2023-07-25','2023-10-24', '2024-01-30', '2024-04-25', '2024-07-30','2024-10-31']
tesla_speech_dates = ['2020-04-29', '2020-07-22', '2020-10-21', '2021-01-27', '2021-04-26', '2021-07-26', '2021-10-20', '2022-01-26', '2022-04-20', '2022-07-20', '2022-10-19', '2023-01-25', '2023-04-19', '2023-07-19', '2023-10-18', '2024-01-24', '2024-04-23', '2024-07-23', '2024-10-23', '2025-01-29']
apple_speech_dates = [
    '2020-01-28', '2020-04-30', '2020-07-30', '2020-10-29', '2021-01-27', '2021-04-28', '2021-07-27', '2021-10-28', '2022-01-27', '2022-04-28', '2022-07-28', '2022-10-27',  '2023-02-02', '2023-05-04', '2023-08-03', '2023-11-02',  '2024-02-01', '2024-05-02', '2024-08-01', '2024-10-31'
]
def regression_and_plot(company_name, sentiment_data, stock_prices, speech_dates, ax):  # Added ax parameter
    speech_dates = pd.to_datetime(speech_dates)
    filtered_prices = stock_prices[stock_prices.index.isin(speech_dates)]

    if filtered_prices.empty:
        print(f"Warning: No matching stock price found for {company_name} on the specified speech dates.")
        return None
    sentiment_data = sentiment_data[:len(filtered_prices)]
    prices_for_regression = filtered_prices['Price'].values

    slope, intercept, r_value, p_value, std_err = stats.linregress(sentiment_data, prices_for_regression)

    regression_line = slope * np.array(sentiment_data) + intercept

    ax.plot(sentiment_data, prices_for_regression, 'o', label=f'{company_name} Stock Price',
             markersize=8, markerfacecolor='skyblue', markeredgecolor='navy')
    ax.plot(sentiment_data, regression_line, label='Regression Line', color='coral', linewidth=2)
    ax.set_title(f'{company_name} Stock Price vs. Sentiment Score', fontsize=16, fontweight='bold')
    ax.set_xlabel('Sentiment Score', fontsize=12)
    ax.set_ylabel('Stock Price', fontsize=12)
    ax.legend(fontsize=10)
    ax.grid(True, linestyle='--', alpha=0.7)


    return {
        'Company': company_name,
        'Slope': slope,
        'Intercept': intercept,
        'R-squared': r_value**2,
        'p-value': p_value,
        'Standard Error': std_err,
        't-statistic': slope / std_err,
    }


fig, axes = plt.subplots(1, 3, figsize=(24, 6))

results_table = []
for i, (company_name, sentiment_data, stock_prices, speech_dates) in enumerate([
    ('Microsoft', microsoft_sentiment, microsoft_prices, microsoft_speech_dates),
    ('Tesla', tesla_sentiment, tesla_prices, tesla_speech_dates),
    ('Apple', apple_sentiment, apple_prices, apple_speech_dates),
]):
    results = regression_and_plot(company_name, sentiment_data, stock_prices, speech_dates, axes[i]) # Pass axes[i]
    if results:
        results_table.append(results)

plt.tight_layout()
plt.show()

results_table = []
for i, (company_name, sentiment_data, stock_prices, speech_dates) in enumerate([ # Added enumerate for 'i'
    ('Microsoft', microsoft_sentiment, microsoft_prices, microsoft_speech_dates),
    ('Tesla', tesla_sentiment, tesla_prices, tesla_speech_dates),
    ('Apple', apple_sentiment, apple_prices, apple_speech_dates),
]):

    results = regression_and_plot(company_name, sentiment_data, stock_prices, speech_dates, axes[i % 2])
    if results:
        results_table.append(results)

plt.tight_layout()
plt.show()

results_table = []
for company_name, sentiment_data, stock_prices, speech_dates in [
    ('Microsoft', microsoft_sentiment, microsoft_prices, microsoft_speech_dates),
    ('Tesla', tesla_sentiment, tesla_prices, tesla_speech_dates),
    ('Apple', apple_sentiment, apple_prices, apple_speech_dates),
]:
    results = regression_and_plot(company_name, sentiment_data, stock_prices, speech_dates, axes[0]) # Pass axes[0] for now, you'll need to adjust this if you want to plot on a specific subplot
    if results:
        results_table.append(results)

results_df = pd.DataFrame(results_table)

if not results_df.empty:
    print("\n Summary Table of Regressions:")
    print(results_df[['Company', 'R-squared', 'Slope', 'p-value',
                     'Standard Error', 't-statistic']].to_string(index=False))
else:
    print("Warning: No regression results were generated.")
```
